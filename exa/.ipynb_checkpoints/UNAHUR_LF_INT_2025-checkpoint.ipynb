1. **[1 punto]** Explíquese $L^* = \{\lambda\} \cup L^+$

El cierre de Kleene de un lenguaje L, que se escribe L*, es el conjunto de todas las cadenas que se pueden formar concatenando elementos de L, y además incluye la cadena vacía. En cambio, el cierre positivo L⁺ es lo mismo pero sin la cadena vacía. Por eso se cumple que L*= {λ} ∪ L⁺

2. **[1 punto]** Explíquese LR $\subset$ LIC

Los lenguajes regulares son más simples y se pueden describir con autómatas finitos o expresiones regulares. Los lenguajes independientes de contexto son más generales y se describen con gramáticas independientes de contexto o autómatas con pila. Como los lenguajes regulares son un caso particular, siempre están incluidos dentro de los independientes de contexto

6. **[1 punto]** Respóndase: Como reconocer un LR infinito si se tiene un número finito de estados en un AF

Un lenguaje regular infinito se reconoce porque el autómata finito que lo acepta tiene ciclos en su grafo de estados. Esos ciclos permiten generar cadenas cada vez más largas, ya que se pueden recorrer indefinidamente
