# Trabajo Práctico Especial 2026

## Objetivo

* Diseñar las GIC en formato BNF de los componentes y sentencias de un lenguaje de programación, siguiendo las siguientes premisas:

## GIC de Componentes del lenguaje:

* **id** = identificador que comienza con una letra y puede seguir con letras, dígitos o guiones. No puede tener dos guiones seguidos y no puede terminar con guión.

* **constd** = constante decimal entera que es una secuencia de dígitos del 0 al 9, no puede comenzar con cero y puede ser positiva o negativa. Las constantes positivas no tienen signo, las negativas tienen el símbolo ‘-’ antes de los dígitos. No es parte del lenguaje la cadena 000 o -000. Ejemplos: 1234, -33, 456, -1, 0.

* **constbin** = constante binaria que comienza con 0 sigue con b y luego una secuencia de 0´s y 1´s. A continuación de la b debe haber al menos un 0 o un 1.

* **Nota**: usar estas reglas en la GIC de sentencias como si fueran terminales **id, constd y constbin.**

## GIC de Sentencias del lenguaje:

* **Ciclo de repetición**: Implementación de un ciclo de repetición(while, for, do while, etc).

* **Decisión**: Implementación de sentencia de decisión. Considere que se pueda anidar sentencias de decisión.

* **Nota**: Defina la estructura y los símbolos terminales que necesite de ambas sentencias. 

* **Asignación**: Implementación de sentencia de asignación simple o múltiple, de la forma: id = expresión o id = id = id = expresión. Expresión es una expresión aritmética, con sumas, restas, multiplicaciones y divisiones, que debe respetar la precedencia matemática. El símbolo = y los símbolos +, -, * , /  y paréntesis son terminales. Se pueden hacer operaciones con constantes decimales, constantes binarias y variables.
    * **Ejemplo de expresiones aritméticas**:
    * id + constd * constbin
    * (id + constd) * constbin

* **Salida**: Las salidas se implementan con en el siguiente formato:
    * **display{ (x1,y1), (x2, y2), .. (xm ym) ; obj1, obj2,….objm}**
    * La palabra reservada **display** y a continuación una lista encerrada entre llaves de una lista no vacía de coordenadas (xi, yi) y a continuación una lista no vacía de objetos obji. 
    * Las coordenadas (xi, yi) son pares de constantes numéricas decimales.
    * Los objetos obji son constantes numéricas o variables. 
    * Las llaves, los paréntesis, la coma, punto y coma y la palabra **display** son terminales.
    * **En la sentencia debe haber la misma cantidad de elementos (xi, yi) que de elementos obji.**
        * Ejemplo:  **display { (constd,constd), (constd, constd)  ;  constbin, id }**

* **Condiciones**: Las condiciones para un constructor de ciclos o de selección deben ser comparaciones binarias simples que pueden estar ligadas por un único conector lógico (**and** u **or**):
    * Ejemplos:
        * expresión **<** expresión
        * expresión **>=** expresión **and** expresión **==** expresión
        * expresión **<** expresión **or** expresión **<=** expresión
    * **Los paréntesis son terminales.**
    * **Los operadores de comparación y lógicos son terminales.**

* **Programa**:  El programa tendrá la siguiente estructura:

    ```
       begin
       Declaración de variables
       Lista de sentencias
       end
    ```

    * Las palabras **begin** y **end**  son terminales.
    * **Declaración de variables**: Pueden ser varias líneas de la forma:
        * tipo : id1, id2, …idn
        * Donde tipo puede ser **integer, float o double**, que son terminales.
    * **Lista de sentencias**: Serán una o más sentencias de las definidas en el lenguaje.

**IMPORTANTE: La Declaración de variables es obligatoria en el programa y debe estar antes de la Lista de sentencias. La Lista de sentencias debe tener al menos una sentencia.**

## Entregables

* Documento digital que contenga:
    * **Las reglas de la gramática** del nuevo lenguaje de programación, teniendo en cuenta las consideraciones.
	* **Una cadena** que pertenezca al lenguaje, es decir, un programa escrito en el lenguaje de programación, que tenga declaración de variables y en la sección del programa tenga al menos tres sentencias de las definidas en el lenguaje.
	* **El árbol de derivación** de la cadena (programa),  para poder verificar que la cadena corresponde al lenguaje.
* Archivo .flap:
    * **Máquina de Turing** que acepte una cadena de la sentencia display. Puede diseñar unicinta o multicinta. Describa la estrategia utilizada.
